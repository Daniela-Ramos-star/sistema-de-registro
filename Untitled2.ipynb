```python
# Cargar los datos
import pandas as pd

url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)

print(df.head())

print(df.info())

print(df.isnull().sum())

df = df.dropna()
```

***Transformación de datos***

```python
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

df.drop(columns=['customer', 'phone', 'internet', 'account'], inplace=True)

df = pd.concat([df, customer_df, phone_df, internet_df, account_df], axis=1)

df.head()
```

```python
import pandas as pd

def comparar_abandono(df, columna_categorica, columna_objetivo='abandono', porcentaje=True):


    if columna_categorica not in df.columns:
        raise ValueError(f"La columna '{columna_categorica}' no existe en el DataFrame.")

    if columna_objetivo not in df.columns:
        raise ValueError(f"La columna '{columna_objetivo}' no existe en el DataFrame.")


    df_filtrado = df[[columna_categorica, columna_objetivo]].dropna()


    valores_unicos = df_filtrado[columna_objetivo].unique()
    if len(valores_unicos) != 2:
        raise ValueError(f"La columna '{columna_objetivo}' debe tener exactamente dos valores (binaria). Encontrado: {valores_unicos}")


    tabla = pd.crosstab(
        df_filtrado[columna_categorica],
        df_filtrado[columna_objetivo],
        normalize='index' if porcentaje else False
    )

    if porcentaje:
        tabla = tabla * 100
        tabla = tabla.round(1)
        tabla.columns = ['% Permanece', '% Abandona']  # Asume 0 = permanece, 1 = abandona
    else:
        tabla.columns = ['Permanece', 'Abandona']

    return tabla
```
